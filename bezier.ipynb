```python
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
```

```python
ecken = np.array([
                  [255,0,0],
                  [0,255,0],
                  [0,0,255],
                ])

def bezier(t, ecken):
    step1 = ecken[0] + t* (ecken[1]-ecken[0])
    step2 = ecken[1] + t* (ecken[2]-ecken[1])
    step3 = step1 + t* (step2-step1)

    return step3
```

```python
vals = bezier(np.expand_dims(np.linspace(0,1,80), axis=-1), ecken)
img = np.array([np.outer(vals[:,0], np.ones(30)), np.outer(vals[:,1], np.ones(30)), np.outer(vals[:,2], np.ones(30))])
img = img.transpose()
print(img.shape)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(vals[:,0], vals[:,1], vals[:,2])
plt.xlim([0,255])
plt.ylim([0,255])
plt.title("Bezierkurve")
plt.show()

plt.figure()
plt.imshow(np.int_(img))
plt.title("Farbverlauf")
plt.show()
```
