```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import io

image_url = "https://www.ln-online.de/resizer/eNn4ji5O0DnQg_pzMPe_x4eYFRI=/1200x675/filters:quality(70)/cloudfront-eu-central-1.images.arcpublishing.com/madsack/UPSKB72653KNZQBTO4KYPWUQ2U.jpg"
img = io.imread(image_url)/255

plt.figure()
plt.imshow(img)
plt.show()

print(img.shape)
```

```python
img_gray = img@np.array([0.2126, 0.7152, 0.0722])
img_gray = np.repeat(img_gray[:,:,None], 3, axis=-1)
print(img_gray.shape)

base = np.outer(np.ones(img.shape[0]), np.linspace(-0.5,1.5, num=img.shape[1]))
base = np.clip(base, a_min=0, a_max=1)[:,:,None]

plt.figure(figsize=[8,10])
plt.subplot(3,1,1)
plt.imshow(img)
plt.subplot(3,1,2)
plt.imshow(img_gray)
plt.subplot(3,1,3)
plt.imshow(base*img_gray+(1-base)*img)
plt.show()
```
