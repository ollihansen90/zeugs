```python
import torch
import matplotlib.pyplot as plt
```

```python
N = 1000
data = torch.rand(N,2)
label = torch.ones(N)
label[:N//2] = -1
label = label[torch.randperm(N)]

plt.figure()
plt.plot(data[label==1, 0], data[label==1,1], "b.")
plt.plot(data[label==-1, 0], data[label==-1,1], "r.")
plt.show()

lr1 = 1e-3
lr2 = 1e-3
n_epochs = 10

for epoch in range(n_epochs):
    perm = torch.randperm(N)
    data = data[perm]
    label = label[perm]
    for l, point in zip(label,data):
        data[label==l] += lr1*(point-data[label==l])
        data[label==-l] -= lr2*(point-data[label==-l])

    plt.figure()
    plt.plot(data[label==1, 0], data[label==1,1], "b.")
    plt.plot(data[label==-1, 0], data[label==-1,1], "r.")
    plt.show()
```
