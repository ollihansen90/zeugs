```python
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

```

```python
# https://matplotlib.org/stable/tutorials/colors/colormaps.html
CM = cm.get_cmap('viridis') # z.B. ['viridis', 'plasma', 'inferno', 'magma', 'cividis']
```

```python
base = np.arange(0,255)
colors = CM(base)[:,:-1]
colors = np.repeat((colors[None,:,:]), 255, axis=0)
print(colors.shape)

plt.figure()
plt.imshow(colors)
plt.show()
```

```python
base = np.arange(0,255)
colors = CM(base)[:,:-1]
print(colors.shape)

plt.figure()
plt.plot(colors[:,0], "r")
plt.plot(colors[:,1], "g")
plt.plot(colors[:,2], "b")
plt.show()
```
