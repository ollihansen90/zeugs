```python
import numpy as np
import matplotlib.pyplot as plt

base = np.linspace(-1,1, num=1000)
f = lambda a,freq, x: a*np.array([np.sin(freq*2*np.pi*x), np.cos(freq*2*np.pi*x)])

y = f(1,1,base)
for i in range(2,10):
    y += f(np.sqrt(i),i,base)
plt.figure()
plt.plot(y[0], y[1])
plt.show()
```
