## Rule 110

```python
import numpy as np
import matplotlib.pyplot as plt

def rulify(num):
    out = np.zeros(8)
    out[0] = num%2
    for ii in range(7):
        i = 7-ii
        out[i] = int(num>=2**i)
        num -= out[i]*2**i
    return out.astype(int)

n_epochs = 200
n_felder = 200
img = np.zeros([n_epochs, n_felder])
img[0,n_felder//2] = 1
#img[0] = np.random.rand(n_felder)>0.5
rules = rulify(30) # 30, 54, 73, 84, 90, 110, 184,
print(rules)

for epoch in range(1,n_epochs):
    update = np.roll(img[epoch-1], 1)*4+img[epoch-1]*2+np.roll(img[epoch-1], -1)
    img[epoch] = rules[update.astype(int)]

plt.figure()
plt.imshow(img)
plt.show()
```

## Noise2Map

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.auto import trange
from PIL import Image

def conv2d(img, kernel):
    ii, jj, nn = img.shape
    ki, kj = kernel.shape
    out = np.zeros_like(img)
    for n in range(nn):
        im = img[:,:,n]
        img_padded = pad(im)
        for i in range(ki):
            for j in range(kj):
                #print(img_padded[i:ii+i, j:jj+j].shape)
                out[:,:,n] += img_padded[i:ii+i, j:jj+j]*kernel[i,j]
    return out

def pad(img, mode="zeros"):
    out = np.zeros([img.shape[0]+2, img.shape[1]+2])
    out[1:-1,1:-1] = img
    return out

def iteration(img):
    out = conv2d(img, kernel=np.array([[1,1,1], [1,0,1], [1,1,1]]))
    return out>=4

h, w, n = 300, 300, 50
thresh = 0.65
vid = np.zeros([n, h, w, 3])
vid[0] = np.random.rand(h,w, 3)>thresh

for epoch in trange(1, n):
    vid[epoch] = iteration(vid[epoch-1])

# plt.figure(figsize=[10,10])
# for epoch in range(9):
#     plt.subplot(3,3,epoch+1)
#     plt.imshow(vid[epoch])
# plt.show()

gif = Image.fromarray(np.uint8(vid[0]*255))
gif.save("hallo.png")
rest = [Image.fromarray(np.uint8(vid[i]*255)) for i in range(1,len(vid))]
gif.save("test.gif", save_all=True, append_images=rest, duration=100, loop=20)

from IPython.core.display import display, HTML
from IPython.core.display import Image as htmlimg
display(htmlimg(filename="test.gif"))
```
